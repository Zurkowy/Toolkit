```python
from selenium import webdriver as wd
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
from openpyxl import load_workbook
import pandas as pd
import time, random, re
```

```python
def find(xpath):
    global browser
    elements = browser.find_elements_by_xpath(xpath)
    while len(elements) == 0:
        time.sleep(0.5)
        elements = browser.find_elements_by_xpath(xpath)
    return elements[0]

def check_exists_by_xpath(xpath):
    try:
        browser.find_element_by_xpath(xpath)
    except NoSuchElementException:
        return False
    return True
```

```python
url = "https://linkdb.kantar.co.uk/logon.aspx?ReturnUrl=%2f"

no_client_list = []

input_data = {}

input_df = pd.read_excel("x1_input_data.xlsx")

#Creating dict from df
for index, row in input_df.iterrows():
    row["client"] = row["client"].strip()
    if row["client"] not in input_data:
        input_data[row["client"]] = []
        input_data[row["client"]].append(row["job_id"])
    else:
        input_data[row["client"]].append(row["job_id"])

for key in input_data:
    print(f"Klient: {key}, \nLiczba reklam: {len(input_data[key])}, \nLista reklam: {input_data[key]}\n\n")

browser = wd.Chrome()
browser.get(url)

#Wait for page to load
WebDriverWait(browser, 30).until(
    EC.presence_of_element_located(
        (By.XPATH,'//*[@id="ctl00_DefaultBreadCrumb"]/span[3]')))

#Compare Norms Click
find('//*[@id="ctl00_ancNorms"]/a/span').click() #Norms Click
find('//*[@id="ctl00_DefaultContent_gv_NormsTypes_ctl03_btnNormsType"]').click()

#Wait for Compare Norms Page to load
WebDriverWait(browser, 30).until(
    EC.presence_of_element_located(
        (By.XPATH,'//*[@id="ctl00_DefaultBreadCrumb"]/span[5]')))

#Choosing Report Type
report_type_selection_menu = browser.find_element_by_xpath('//*[@id="ctl00_DefaultContent_ddlReportType"]')
for option in report_type_selection_menu.find_elements_by_tag_name('option'):
    if option.text == 'Raw Scores':
        option.click()    

time.sleep(5)

#Choosing Solution
WebDriverWait(browser, 30).until(
    EC.element_to_be_clickable(
        (By.XPATH, '//*[@id="select2-ctl00_DefaultContent_ddlSolution-container"]')))
find('//*[@id="select2-ctl00_DefaultContent_ddlSolution-container"]').click()
find('/html/body/span/span/span[1]/input').send_keys("Link for TV")
find('/html/body/span/span/span[1]/input').send_keys(Keys.ENTER)

time.sleep(5)

#Choosing Market
WebDriverWait(browser, 30).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="select2-ctl00_DefaultContent_selCountries-container"]')))
find('//*[@id="select2-ctl00_DefaultContent_selCountries-container"]').click()
find('/html/body/span/span/span[1]/input').send_keys("Poland")
find('/html/body/span/span/span[1]/input').send_keys(Keys.ENTER)

#Choosing Data Methodology
WebDriverWait(browser, 30).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="select2-ctl00_DefaultContent_cbLstDCM-container"]')))
find('//*[@id="select2-ctl00_DefaultContent_cbLstDCM-container"]').click()
find('/html/body/span/span/span[1]/input').send_keys("Online")
find('/html/body/span/span/span[1]/input').send_keys(Keys.ENTER)

# #Choosing Measure
# measure_list = []

# for measure in measure_group_name_list:
#     find('//*[@id="ctl00_DefaultContent_reportFilters"]/div[5]/div[2]/span/span[1]/span/ul/li/input').send_keys(measure)
#     find('//*[@id="ctl00_DefaultContent_reportFilters"]/div[5]/div[2]/span/span[1]/span/ul/li/input').send_keys(Keys.ENTER)

#Choosing Measure Groups
# measure_group_list = ["Topline"]
measure_group_list = ["Topline", "Campaign", "Celebrity", "Communication", "Compulsory",
                      "Key Ratings", "Music","New Emotions", "Open Ends", "Optional",
                      "Persuasion - CPG", "Persuasion - Long Term", "persuasion - Services",
                      "Sub brand Persuasion - CPG", "Sub brand Persuasion - Long Term",
                      "Sub brand Persuasion - Services", "Dashboard", "MDF (Meaningfully Different Framework)",
                      "Affectiva", "Message Check", "Intuitive Associations", "Behavioural", "Calculated Scores",
                      "Emoticons", "Stereotype", "ANA – GEM – Gender Equality Measure", "UA – UM – Unstereotype Metric",
                      "Polarisation", "Brand Purpose"
                     ]

for measure_group in measure_group_list:
    find('//*[@id="ctl00_DefaultContent_reportFilters"]/div[6]/div[2]/span/span[1]/span/ul/li/input').send_keys(measure_group)
    find('//*[@id="ctl00_DefaultContent_reportFilters"]/div[6]/div[2]/span/span[1]/span/ul/li/input').send_keys(Keys.ENTER)
    
#Choosing Job ID/NUMBER
report_type_selection_menu = browser.find_element_by_xpath('//*[@id="ctl00_DefaultContent_updAd1Search"]')
for option in report_type_selection_menu.find_elements_by_tag_name('option'):
    if option.text == 'Job Id':
        option.click() 

for client_name in input_data:
    
    #Creating list for no ads
    no_ad_list = []
    ad_counter = 0
        
    #Creating loading_screen variable
    ignored_exceptions=(NoSuchElementException, StaleElementReferenceException,)
    loading_screen = WebDriverWait(browser, 5, ignored_exceptions = ignored_exceptions).until(
        EC.presence_of_element_located(
            (By.XPATH, '//*[@id="ctl00_DefaultContent_UpdateProgressAd1"]/div/div/img')))
    
    #Check if "loading screen" is displayed
    try:
        while loading_screen.is_displayed():
            time.sleep(1)
        else:
            pass
    except:
        pass
    
    #Cleaning client_name to enable create a file
    client_file_name = re.sub('[^a-zA-Z0-9 ,.-=!@#$%^&()\n]', '_', client_name)
    ExcelWriter = pd.ExcelWriter(str(client_file_name)+".xlsx")

    for job_id in input_data[client_name]:
        
        #Choosing client
        browser.find_element_by_xpath("//*[@id='ctl00_DefaultContent_trwHierarchyt"+str(client_list.index(client_name))+"']/input").click()
        
        ad_counter += 1
        
        print(f"Obecny status:\nKlient:{client_name}\nJob_ID:{job_id}\t{ad_counter}/{len(input_data[client_name])}")
        
        #Creating loading_screen variable
        ignored_exceptions=(NoSuchElementException, StaleElementReferenceException,)
        loading_screen = WebDriverWait(browser, 5, ignored_exceptions = ignored_exceptions).until(
            EC.presence_of_element_located(
                (By.XPATH, '//*[@id="ctl00_DefaultContent_UpdateProgressAd1"]/div/div/img')))    
        
        #Choosing Job_ID/Job_Number
        find('//*[@id="ctl00_DefaultContent_txtAdSearch1"]').clear()
        find('//*[@id="ctl00_DefaultContent_txtAdSearch1"]').send_keys(job_id)
        WebDriverWait(browser, 30).until(
            EC.element_to_be_clickable(
                (By.XPATH, '//*[@id="ctl00_DefaultContent_lnkBtnAdSearch1"]')))
        find('//*[@id="ctl00_DefaultContent_lnkBtnAdSearch1"]').click()
        
        time.sleep(3)     
        
        #Check if "loading screen" is displayed
        
        try:
            while loading_screen.is_displayed():
                time.sleep(1)
            else:
                pass
        except:
            pass
        
        #Check if "No AD Info" is showing up
        try:
            if check_exists_by_xpath('//*[@id="ctl00_DefaultContent_divAd1NoData"]/em'):
                no_ad = WebDriverWait(browser, 5).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="ctl00_DefaultContent_divAd1NoData"]/em')))
                no_ad_list.append(job_id)
                continue
            else:
                pass
        except:
            pass
        
        #Choosing Client + Check if "loading screen" is displayed
        try:
            while loading_screen.is_displayed():
                time.sleep(1)
            else:
                pass
        except:
            pass
        
        #Checking if ad client is chosen
        try:
            if check_exists_by_xpath('//*[@id="ctl00_DefaultContent_divAdSearch1"]/span/em'):
                no_client = WebDriverWait(browser, 5).until(
                    EC.presence_of_element_located(
                        (By.XPATH,'//*[@id="ctl00_DefaultContent_divAdSearch1"]/span/em')))
                #Waiting and choosing Client
                client_xpath = "//*[@id='ctl00_DefaultContent_trwHierarchyt"+str(table_counter)+"']/input"
                WebDriverWait(browser, 30).until(
                    EC.element_to_be_clickable(
                        (By.XPATH, client_xpath)))
                find(client_xpath).click()
                find('//*[@id="ctl00_DefaultContent_lnkBtnAdSearch1"]').click()
                pass
            else:
                pass
        except:
            pass
        
        #Waiting for Job ID to show up
        WebDriverWait(browser, 30).until(
            EC.presence_of_element_located(
                (By.XPATH,'//*[@id="ctl00_DefaultContent_gvAds1"]/tbody/tr[1]')))
        
        #Checking ID
        if find('//*[@id="ctl00_DefaultContent_gvAds1_ctl02_lbljobId"]').text != job_id:
            time.sleep(1)
        else:
            pass
        
        #Marking AD
        WebDriverWait(browser, 30).until(
        EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="ctl00_DefaultContent_gvAds1_ctl02_chkAd1"]')))
        find('//*[@id="ctl00_DefaultContent_gvAds1_ctl02_chkAd1"]').click()
        
        #Generating Report (click)
        find('//*[@id="ctl00_DefaultContent_btnGenReport"]').click()
        
        #Waiting for table to show up
        WebDriverWait(browser, 120).until(
            EC.presence_of_element_located(
                (By.XPATH,'//*[@id="ctl00_DefaultContent_repeaterNorms_ctl00_lblMeasureName"]')))
        
        #Scraping + cleaning table
        source_data = bs(browser.page_source,'html')
        tables = source_data.find_all('table')
        main_table = tables[4]
        table_data = [[str(cell.text).strip() for cell in row.find_all(["th","td"])] for row in main_table.find_all("tr")]
        
        #Table to DF + export
        output_data = pd.DataFrame(table_data)
        
        output_data.columns = output_data.iloc[0]
        output_data = output_data.drop(output_data.index[0])
        output_data = output_data.drop([""], axis = 'columns')
        output_data["Job_id"] = job_id
        
        output_data.to_excel(
            ExcelWriter,
            index = False,
            sheet_name = str(job_id))
        
        #Back to previous page (Configure click)
        find('//*[@id="ctl00_DefaultContent_btnConfigure"]').click()
        
        #Wait for "Search for Ad1" to show up
        WebDriverWait(browser, 30).until(
        EC.presence_of_element_located(
            (By.XPATH,'//*[@id="ctl00_DefaultContent_divFilter"]/div[1]/div[1]/div')))
        
    ExcelWriter.save()
    
    if len(no_ad_list) > 0:
        No_ad_frame = pd.DataFrame(
            data = {"Job_ID": no_ad_list })
        
        No_ad_frame.to_csv(
            client_name+"_NO_AD_LIST.csv",
            sep = ',',
            index = False)
    
    print(f"\t\tZakończono:\n{client_name}\nBrakowało: {len(no_ad_list)}\n\n")
    
if len(no_client_list) > 0:
    No_client_frame = pd.DataFrame(
        data = {"Client_Name": no_client_list })

    No_client_frame.to_csv(
        "No_client_list.csv",
        sep = ',',
        index = False)
    
ExcelWriter.save() 
ExcelWriter.close()
browser.quit()

print("Zakończono pomyślnie.")
```
